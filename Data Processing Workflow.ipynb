## Remove points outside Leb

```python
import geopandas as gpd
```

```python
leb_path = r"C:\Users\User\OneDrive\Desktop\Administrative_Divisions\Administrative_Divisions.gdb"

leb = gpd.read_file(leb_path, layer="Lebanon")
```

```python
gdf = gpd.GeoDataFrame(
    df,
    geometry=gpd.points_from_xy(df.lon, df.lat),
    crs="EPSG:4326"
)
```

```python
gdf = gdf.to_crs(leb.crs)
```

```python
# spatial clip
gdf_clipped = gpd.sjoin(gdf, leb, predicate="within", how="inner")
```

```python
print("Before:", len(gdf))
print("After:", len(gdf_clipped))
print("Removed:", len(gdf) - len(gdf_clipped))
```

```python
# drop join cols
gdf_clipped = gdf_clipped.drop(columns=["index_right"])
```

## Data Enrichment

### Decode Coded Fields

```python
radio_map = {
    "GSM": "2G",
    "UMTS": "3G",
    "LTE": "4G"
}

net_map = {
    1: "Alfa",
    3: "Touch"
}
```

```python
gdf_clipped["network_type"] = gdf_clipped["radio"].map(radio_map)
gdf_clipped["provider"] = gdf_clipped["net"].map(net_map)
```

### Add Gov Name

```python
admin_path = r"C:\Users\User\OneDrive\Desktop\Administrative_Divisions\Administrative_Divisions.gdb"

gov = gpd.read_file(admin_path, layer="Admin1")

gov = gov[['Name_En', 'geometry']] 
```

```python
gdf_joined = gpd.sjoin(
    gdf_clipped,
    gov,
    predicate="within",
    how="left"
)
```
